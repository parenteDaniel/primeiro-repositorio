```python
import random
import math

def gerar_numeros_aleatorios_pares_impares_ordenados(qtd_pares, qtd_impares, exclusoes):
    numeros_pares = set()
    numeros_impares = set()

    # Gerar números pares
    while len(numeros_pares) < qtd_pares:
        numero_aleatorio = random.randrange(2, 61, 2)  # Escolhe apenas números pares
        if numero_aleatorio not in exclusoes:
            numeros_pares.add(numero_aleatorio)

    # Gerar números ímpares
    while len(numeros_impares) < qtd_impares:
        numero_aleatorio = random.randrange(1, 61, 2)  # Escolhe apenas números ímpares
        if numero_aleatorio not in exclusoes:
            numeros_impares.add(numero_aleatorio)

    # Unir os conjuntos e ordenar
    numeros_ordenados = sorted(list(numeros_pares | numeros_impares))

    return numeros_ordenados


def calcula_distribuicao_perfeita(qt_dezenas_jogo):

  lista_numeros = list(range(1, 61))

  if (qt_dezenas_jogo % 2 == 0):
    qt_selecionados = qt_dezenas_jogo / 2
    param_do_meio = 0
  else:
    qt_selecionados = math.floor(qt_dezenas_jogo / 2)
    param_do_meio = 29

  print("Selecionados para slice:", qt_selecionados)

  soma_dos_primeiros = sum(lista_numeros[:int(qt_selecionados)])

  qt_selecionados = qt_selecionados * -1

  soma_dos_ultimos = sum(lista_numeros[int(qt_selecionados):])

  print("Soma dos primeiros:",soma_dos_primeiros)
  print("Soma dos ultimos:",soma_dos_ultimos)

  soma_total = soma_dos_primeiros + param_do_meio + soma_dos_ultimos

  return soma_total

# Algortimo
# 01-03-23-27-47-57
exclusoes = [1,3,23,27,47,57]  # Substitua isso pelos números que você quer excluir
qt_dezenas = 6
qt_jogos = 2
param_lim = 0.05

print ("Quantidade de dezenas:",qt_dezenas)
print ("Quantidade de jogos:",qt_jogos)

#calcula distribuicao perfeita de acordo com numero de dezenas
param_distr = calcula_distribuicao_perfeita(qt_dezenas)
print ("Distribuição Perfeita:",param_distr)

param_distr_min = param_distr - (param_distr * param_lim)
param_distr_max = param_distr + (param_distr * param_lim)

jogos_gerados = 1
soma_numeros = 0

while (jogos_gerados <= qt_jogos):

  #determinar qtos pares e qtos impares terao os jogos
  if (qt_dezenas % 2 != 0):
    numero_de_pares = random.randint((qt_dezenas-1)/2, ((qt_dezenas-1)/2)+1)
  else:
    numero_de_pares = random.randint((qt_dezenas / 2)-1, (qt_dezenas / 2)+1)

  numero_de_impares = qt_dezenas - numero_de_pares
  print ("Numero de Impares:",numero_de_impares, ",jogo: ", jogos_gerados)
  print ("Numero de Pares:", numero_de_pares, ",jogo: ", jogos_gerados)

  while ((soma_numeros < param_distr_min) or (soma_numeros > param_distr_max)):
    numeros_aleatorios = gerar_numeros_aleatorios_pares_impares_ordenados(numero_de_pares, numero_de_impares, exclusoes)
    soma_numeros = sum(numeros_aleatorios)

  print("Jogo número:", jogos_gerados, "Dezenas geradas:", numeros_aleatorios)
  print("Jogo número:", jogos_gerados, "Soma Dezenas geradas:", soma_numeros)
  soma_numeros = 0

  jogos_gerados = jogos_gerados+1
```
